```python
from ultralytics import YOLO

# выбор загружаемой модели
model = YOLO('E:/diplom/v1/runs/detect/train21/weights/best.pt')

# обучение модели  
results = model.train(data="yolo_dataset/config.yaml", epochs=10, imgsz=640, batch=32, patience=5, verbose=True, save=True )

# epochs - продолжительность обучения, чем больше, тем дольше
# batch - сколько изображений обрабатывается за обновление/тик. Autobatch - выставляет кол-во автоматически под gpu
# patience - Количество эпох которые будут без улучшения показателей для того чтобы обучение остановилось
# verbose - подробные данные во время обучения
# save - возможность сохранять результаты, даже если обучение остановилось или прекратилось, в последствии можно продолжать обучение после оставной точки
# verbose - подробные данные во время обучения
# verbose - подробные данные во время обучения
```

```python
# загрузка модели 
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import torch
import numpy as np

model = YOLO('runs/detect/train18/weights/best.pt')
img = cv2.imread('images/3c7b1041-1e18-459d-95bf-3a92b118956e.jpg')
print(img.shape)

# Обнаружение объекта
results = model(img, imgsz=640, iou=0.4, conf=0.5, verbose=True)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

annotated_frame = results[0].plot()
annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)
plt.imshow(annotated_frame)
plt.show()
```
